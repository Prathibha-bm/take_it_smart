```python
from sklearn.linear_model import LogisticRegression
import numpy as np

data = np.array([
    [25,120,85],#[age,bp,sugar]
    [45,140,150],
    [50,150,180],
    [30,130,95],
    [60,160,200],
    [35,125,90],
    [55,155,170],
    [40,135,110]
])
y = np.array([0,1,1,0,1,0,1,0])

model = LogisticRegression()
model.fit(data,y)

age = int(input("Enter age: "))
bp = int(input("Enter blood pressure: "))
sugar = int(input("Enter sugar level: "))

prediction = model.predict([[age,bp,sugar]])

if prediction[0] == 1:
    print("Disease Risk Detected")
else:
    print(" No Disease Risk")
```
