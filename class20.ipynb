```python
#pridict the height based on the age

import numpy as np
from sklearn.linear_model import LogisticRegression
import pandas as pd

#data
ages=np.array([2,5,7,10,12,14,16,18,20]).reshape(9,1)
height=np.array([74,84,75,110,115,105,130,150,145])
print(ages,height)

hclass=[]

for i in height:
  if i<100:
    hclass.append(0)
  elif i<=130:
   hclass.append(1)
  else:
   hclass.append(2)

hclass=np.array(hclass)
#model
model=LogisticRegression(multi_class='multinomial')
model.fit(ages,height)


import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.scatter(ages,height,c=hclass,cmap='viridis',s=100)
plt.xlabel('age')
plt.ylabel('height')
plt.title('Realistic age vs height')
plt.xlabel('height')
plt.title('realistic age vs height')
plt.colorbar()
plt.show()

#decision boundry
age_range=np.linspace(4,19,200).reshape(-1,1)
pre=model.predict(age_range)
for i in range(len(pre)-1):
  if pre[i]!=pre[i+1]:
    print('boundry near age',age_range[i])
```
