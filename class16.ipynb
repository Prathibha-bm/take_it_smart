```python
import pandas as pd
df=pd.read_csv('/content/Titanic-Dataset.csv')
df.head()

print(df.info())
print(df.describe())

df.rename(columns={'Sex':'gender'},inplace=True) #inplace true without this it does't make changes
df.columns

print(df.isnull().sum())

#problem 1
x=df.groupby('gender')['Survived'].sum()
print('\n1) People servived :\n',x)

#problem 2
df.groupby('Pclass')['Survived'].mean()
print('\n 2)\n',x)

#problem 3
df.groupby('gender')['Age'].median()
print('\n 3)\n',x)

#problrm 4
#df.fillna({'Age':'unknown'},inplace=True)
#print('\n 4) \n',df.isnull().sum())

df['Age'].fillna(df['Age'].median(),inplace=True)
print('\n 4i\n',df.isnull().sum())

#problrm 5
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
print('\n 5)\n',df.isnull().sum())

#problrm 6
df.drop(columns=['Cabin'],inplace=True)
print('\n 6) \n',df.isnull().sum(),'\n \n')

df.to_csv('titanic_cleaned.csv',index=False)

#matplotlib graphs
#how many people survived
#dataset from kaggle - data analysis(sales marketing finace)


import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x='gender',y="Survived",data=df)
plt.title("Servived peoplr based on gender")
plt.show()

 #age distribution

sns.histplot(df['Age'],bins=20)
plt.title('Age Distribution')
plt.show()
```

```python
import pandas as pd

column_names = [
    'Name', 'Age', 'Gender', 'Blood Group', 'Medical Condition',
    'Date of Admission', 'Doctor', 'Hospital', 'Insurance Provider',
    'Billing Amount', 'Room Number', 'Admission Type',
    'Discharge Date', 'Medication', 'Test Results'
]

df = pd.read_csv('/content/healthcare_dataset.csv', names=column_names, skiprows=1)#i spacifi the columns b/z befor pandas reds 1st column as the columns
df.head()

print(df.info())
print(df.describe())

df.rename(columns={'Blood Type':'Blood Group'},inplace=True) #inplace true without this it does't make changes
df.columns

#problem 1
x=df.groupby('Gender')['Admission Type'].sum()
print('\n1) People admission:\n',x)


#problem 2
x=df.groupby('Age')['Billing Amount'].mean()
print('\n 2) Amount spent:\n',x)

#problem 3
df.groupby('Age')['Billing Amount'].median()
print('\n 3) Age \n',x)

df.fillna({'Date of Admission':'unknown'},inplace=True)
print('\n 4)Fill of Date of Admission \n',df.isnull().sum())

#poblem 5
df['Date of Admission'].fillna(df['Date of Admission'].mode()[0],inplace=True)
print('\n 5)\n',df.isnull().sum())

df['Age'].fillna(df['Age'].mode()[0],inplace=True)
print('\n 5i\n',df.isnull().sum())

#problrm 6
df.drop(columns=['Insurance Provider'],inplace=True)
print('\n 6) Droping of Insurence Provider \n',df.isnull().sum(),'\n \n')


df.to_csv('healthcare_cleaned.csv')

import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x='Gender',y="Test Results",data=df)
plt.title("Test result based on gender")
plt.show()

 #age distribution

sns.histplot(df['Age'],bins=20)
plt.title('Age Distribution')
plt.show()

```
