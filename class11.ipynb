1.Create an array and find sum and avg of all the elements in an array


```python
import numpy as np
nums=np.array([1,3,6,3,7,5,8,4,97,4346,567,36,5644])
total=0
count=0
for i in nums:
    total=total + i
    count=count + 1
avg=total/count
print(nums)
print("Total sum :", total)
print("average of all elements :", avg)
```
