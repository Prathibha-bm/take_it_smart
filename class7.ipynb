```python
def add():
  a=int(input("Enter : "))
  b=int(input("Enter : "))
  print(a+b)

def sub():
  a=int(input("Enter : "))
  b=int(input("Enter : "))
  print(a-b)

def divide():
  a=int(input("Enter : "))
  b=int(input("Enter : "))
  print(a/b)

def power():
  a=int(input("Enter : "))
  b=int(input("Enter : "))
  print(a**b)

def reminder():
  a=int(input("Enter : "))
  b=int(input("Enter : "))
  print(a%b)

while True:
  c=input("Enter add/sub/divide/power/remider:").strip().lower()
  if c=='add':
    add()
  elif c=='sub':
    sub()
  elif c=='divide':
    divide()
  elif c=='power':
    power()
  elif c=='reminder':
    reminder()

  else:
   print("invalid Try again!")
```
