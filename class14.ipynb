```python
#pandas : it is a python lybrary used for data manupulation,data analysis  and cleaning, primary handling structured data through FataFrame and Series structures
import pandas as pd

data={
    'age':[23,34,45,56,67],
    'name':['sam','pam','ram','sham','seema'],
    'marks':[248,38,200,549,867],
    'city':['blg','chenni','mum','delhi','pune'],
    'contry':['india','india','None','Eourop','India'],
    'IDHnumber':[None,284,285,286,207]
}

#print(data) dataframe
df=pd.DataFrame(data)
print('\n\nData Frame\n\n',df)

#series
df=pd.Series(data)
print('\n\n1)Series\n\n',df)

x=df.head(2)
print('\n\n2)Head\n\n',x)

x=df.tail(2)
print('\n\n 3)Tail\n\n',x)

```

```python
import pandas as pd
df=pd.DataFrame(data)

print('\n\n1)Columns\n\n',df.columns)
print('\n\n2) Info\n\n',df.info())
print('\n\n3) Describe\n\n',df.describe)
print('\n\n4) iloc\n\n',df.iloc[0:4])
```

```python
#filter data
x=df[df['marks']>100]
print('\n1\n',x)

x=df[df['city']=='delhi']
print('\n2)\n',x)

x=df[(df['city']=='delhi')&(df['marks']>100)]
print('\n3)\n',x)

x=df['marks'].sum()
print('\n4) marks sum:\n',x)

x=df['marks'].count()
print('\n 5)marks count:\n',x)


#adding new column
x=df[' grade']=['b','d','c','a','a']
print('\n6) grade:\n',df)

x=df['Result']=df['marks']>75
print('\n7) Result:\n',df)

#dort_values
x=df.sort_values('marks',ascending=False)
print('\n8) sort_values:\n',x)


#groupby
x=df.groupby('city')['marks'].mean()
print("\n9) groupby\n",x)

#filling missing value

x=df.isnull().sum()
print('\n 10)\n',x)

x=df.fillna('unknown')
print("\n11)fillna \n",x)

x=df.dropna()
print("\n12)dropna\n",x)
```
