```python
def monkey_trouble(a_smile, b_smile):
  if a_smile==b_smile :
    return True
  else:
    return False
```

```python
def sum_double(a, b):
  if a==b:
   return 2*(a+b)
  else :
    return a+b
```

```python
def diff21(n):
  if n>21:
    return 2*(abs(n-21))
  else:
   return abs(n-21)
```

```python
def parrot_trouble(talking, hour):
  if talking==True or (hour<7 and hour>20):
    return True
  else:
    return False
```

```python
def makes10(a, b):
  if a == 10 or b == 10 or a + b == 10:
    return True
  else :
    return False
```

```python
def near_hundred(n):
  if abs(n-100) <=10 or abs(n-200)<=10:
    return True
  else :
    return False
```

```python
def pos_neg(a, b, negative):
  if negative:
        return a < 0 and b < 0
  else:
       return (a < 0 and b > 0) or (a > 0 and b < 0)
```

```python
def not_string(str):
  if str.startswith("not"):
        return str
  else:
        return "not " + str
```

```python
def missing_char(str, n):
   return str[:n] + str[n+1:]
```

```python
def front_back(str):
    if len(str) < 2:
        return str
    return str[-1] + str[1:-1] + str[0]
```

```python
def front3(str):
  front = str[:3]
  return front * 3
```
